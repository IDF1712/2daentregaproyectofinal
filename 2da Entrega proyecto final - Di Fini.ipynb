# ABSTRACT
---

En los últimos años, ha habido un aumento de la investigación y el análisis sobre la importancia de la salud mental para lograr los objetivos mundiales de desarrollo sostenible. La estabilidad mental puede verse afectada por diferentes situaciones: la enfermedad puede afectar la emoción, los pensamientos, la actitud y la toma de decisiones del individuo. Los trastornos mentales son cada vez más comunes para los empleados debido al estrés en su lugar de trabajo. La enfermedad mental puede causar trastorno de personalidad, trastorno de ansiedad, fobias, trastornos psicóticos, depresión, trastornos del estado de ánimo, trastornos alimentarios y algunos más. En este documento, analizamos la causa de los trastornos de salud mental entre los empleados del conjunto de datos de la Encuesta de salud mental en tecnología de Open Sourcing Mental Illness (OSMI). Aquí analizamos la gravedad de la enfermedad mental para los empleados que trabajan en función de diferentes factores o atributos que incluyen trabajo por cuenta propia, historial de salud mental en la familia de los empleados, compañía que ofrece efectos beneficiosos para la salud, si el empleado está en tratamiento por enfermedad mental y mucho más usando Machine. Algoritmos de aprendizaje. El principal objetivo de analizar dichos datos es educar al público sobre las enfermedades mentales en un entorno laboral, por lo que ayuda a reducir los problemas de trastornos mentales. Este documento apoya y aconseja sobre la causa de los comportamientos graves de salud mental y para prevenir cualquier suceso desafortunado debido a diversos factores en un entorno laboral. Por lo tanto, esto proporciona una estimación de cómo los empleados se ven afectados tanto en empresas tecnológicas como no tecnológicas. Este análisis nos lleva a la conclusión de responder preguntas sobre si la ubicación es importante, el número de empleados en una empresa es importante, si la salud mental se tiene en cuenta o no y cómo estos factores externos afectan la salud mental y el bienestar físico en el sectores de trabajo.

#Construcción y evaluación de modelos

##f1_score

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
```

```python
#Se crea la funcion para evaluar cada uno de los modelos
def train_evaluate(model,X_train,y_train,name):
    model.fit(X_train,y_train)
    y_pred = model.predict(X_train)
    f1_train = f1_score(y_train,y_pred)

    #Cross validation
    f1_val = cross_val_score(model,X_train,y_train,scoring='f1',cv=10)
    
    # Devolviendo los scores
    score = pd.DataFrame({'Name' : name ,'F1_score_trainset' : [f1_train], 'F1_score_validationset' : [f1_val.mean()]})
    return score
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty='l1',solver='liblinear')
train_evaluate(log_reg,X_train,y_train,'Logistic Regression')
```

##Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(max_leaf_nodes=4,random_state=42)
train_evaluate(dt_clf,X_train,y_train,'DecisionTreeClassifier')
```

## SVC

```python
from sklearn.svm import SVC
svc_clf = SVC()
train_evaluate(svc_clf,X_train,y_train,'Support Vector Classifier')
```

## Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(random_state=42)
train_evaluate(rnd_clf,X_train,y_train,'RandomForestClassifier')
```

## XGBoost

```python
from xgboost import XGBClassifier
xgb_clf = XGBClassifier(verbosity=0)
train_evaluate(xgb_clf,X_train,y_train,"XG Boost CLASSIFIER")
```
